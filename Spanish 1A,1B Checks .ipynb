```python
import pandas as pd
import numpy as np
```

## Checking that Spanish 1A kids go into 1B

```python
spanish_a = pd.read_csv("Spanish_A_Final_Grades.csv")
spanish_a.head()
```

```python
spanish_b = pd.read_csv("spanish_B_course_requests.csv")
spanish_b.head()
```

```python
merged = spanish_a.merge(spanish_b, how='left', on=["ID"])
merged.count()
```

```python
merged.head()
```

```python
spanish_list = merged.drop(['Crs ID', 'Course ID', 'Student Name', 'Grade','ELP Level', 'SE Level'], axis=1)
spanish_list.head()
```

```python
spanish_list.to_csv("spanish_list.csv")
```

## Checking that All 1B kids have taken 1A 

```python
merged2 = spanish_b.merge(spanish_a, how='left', on=["ID"])
merged2.head()
```

```python
spanish_list2 = merged2.drop(['Crs ID', 'Course ID', 'Name', 'Grade','ELP Level', 'SE Level'], axis=1)
spanish_list2.head()
```

```python
spanish_list2.to_csv("spanish_list2.csv")
```
